## VARIABLES

### What is a Variable?

* A variable is a name (label) that points to an object in memory
* Python does not store values inside variables.
   
**->**  It stores objects in memory, and variables reference them

##### In Simple Terms: 

* In python Variables are used as a storage to store things in 
python.
* You can write anything as a variable name. 

**For example :**
x = 10

10 → object in memory (heap)

x → reference (name) pointing to that object

```python
age = 22
print(age)
```

```python
name ="Kamran Kausher"
print(name)
```

#### How Strings work

* You know what strings are but you must also know string 
take more space than other data types like int, float etc
* This happens because String stores every character with 
their own Unicode
* Unicode is a universal character encoding standard that 
assigns a unique number (code point) to every character, 
regardless of language
* Like “A” unicode is 65, 
you can check them by using ord() function in python and 
convert them back using chr() function.

```python
ord("A")
```

```python
chr(65)
```

#### String Indexing 

* You must have thought there are so many characters in a 
string but can you access everyone.
* Yes thats possible using indexing. Indexing starts from 0 and 
goes till the number of characters you have.
* eg - a = “Hello”  print(a[0]) ==> output - "H"
* There is negative indexing as well and it starts from -1, but 
the starting position is from the back of the string.
* eg - a = “Hello”  print(a[-1]) ==> output - “o”

```python
name = "KAMRAN KAUSHER"
print(name[1])
```

```python
print(name[111])    #error coz. string index out of range
```

```python
print(name[4])
```

```python
print(name[-1])
```

#### String Slicing

* You know how to access characters in string. But there are 
slicing option as well.
* Slicing means cutting out a slice from string and this is also 
done using index values
* eg - a = “hello”  a[1:4:1]  ==> output "ell"
* So here we have start , stop and steps position and keep a 
note if we use stop at 4 it will slice till 3 only. 

```python
# name[start:stop:steps]     , stop = n-1 if we use stop at 4 it will slice till 3 only.
```

```python
name[1:6:1]
```

```python
name[:]
```

```python
name[-1:-15:-1]
```

```python
name[::-1]
```

```python
name[::-2]
```

There are 2 types of conversion **Implicit** and **Explicit** :

* In this python automatically
converts data from one data

* eg - a = 12
  print(a/2)

##### Explicit

**Falsy values**

* 0
* 0.0 
* False

All these values are falsy remaining will be converted to True.

## Python Memory Model

### 🧠 Where things live

* Stack → references (variable names)
* Heap → objects (actual data)

##### Example:

```python
a = 10
b = a
```

Memory view:

Heap:        10

        ↑
            
Stack:  a      b



##### Both a and b point to the same object.

## Mutability vs Immutability

### 🔒 Immutable types

* int
* float
* bool
* string
* tuple

**You cannot change the object itself.**

```python
x = 10
x = x + 1   # new object created
```

### 🔓 Mutable types (preview)

* list
* dict
* set

```python
lst = [1, 2, 3]
lst.append(4)   # same object modified
```

⚠️ This is where ML bugs are born.

### Why This Matters

#### 🚀 ML Performance

* Object creation is expensive
* Immutable objects create new memory
* Vectorized operations avoid overhead

##### 🐛 Bug-Free Code

Example bug:

```python
a = [1, 2]
b = a
b.append(3)
```

Now:

```python
a = [1, 2, 3]  # unexpected
```

This destroys ML pipelines if misunderstood.
