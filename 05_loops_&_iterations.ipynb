# Loops & Iterations

###  Why Loops Exist (First principles)

Computers are fast at repetition.
Loops let you:
* Process large datasets
* Train models over epochs
* Apply transformations repeatedly
* Implement algorithms (DSA)

##### Without loops:

* No ML
* No data pipelines
* No training

***

###  for Loop (Iteration over sequences)

Repeat for EACH item

**🔹 Syntax**

```python
for variable in sequence:
    code_block
```

**🔹 Example**

```python
for i in [1, 2, 3]:
    print(i)
```

```python
name="KAMRAN"
for i in name:
    print(i)
```

```python
L=["KK", 45,True, 60.5 ]
for x in L:
    print(x)
```

**🔹 Using range()**

* Before understanding for loops you should know how range 
function works
* The range() function is used to generate a sequence of 
numbers, which is commonly used in loops
* Syntax of range function is simple range(start, stop, steps)
* you have 3 points from where you want to start, till where 
you want to stop and how many steps you want
* If you don’t mention start point the default value will be 0 .
if you don’t mention the steps the default steps will be 1.
* you have to mention the stop point otherwise the range 
function will not work.

```python
for i in range(5):
    print(i)
```

```python
for x in range(0,21,2):
    print(x)
```

```python
for m in range(11,1,-1):
    print(m)
```

##### Use for when:

* You know what you’re looping over
* The data already exists
* You want clean, safe iteration

##### Examples

* Dataset rows
* Epochs
* Batches
* Tokens
* Characters
  
##### 🧠 Where used in ML

* Iterating over rows
* Iterating over epochs
* Feature loops (basic level)

 ##### Why this matters in ML

* Large epoch counts
* Huge loops
* Memory-sensitive pipelines

### while Loop (Condition-based repetition)

Repeat WHILE a condition is true

**🔹 Syntax**

```python
while condition:
    code_block
```

**🔹 Example**

```python
i = 0
while i < 5:
    print(i)
    i += 1
```

```python
m = 5
while m < 15:
    print(m)
    m +=1
```

#### Loops for strings

Loops for strings work slightly differently. You can iterate 
through a string in two ways :-

**Using index values**

```python
a = "Machine Learning"
for i in range (len(a)):
    print(a[i])
```

**Iterating directly over the string**

```python
for ch in a: 
    print(ch)
```

##### Use while when:

* Loop depends on dynamic state
* Termination is not count-based

##### Real ML examples

* Train until loss converges
* Retry API until success
* Agent runs until stopped
* Streaming dat

##### 🧠 ML usage

* Training until convergence

* Dynamic stopping conditions

### break (Exit loop immediately)

The break statement in Python is used to exit a loop 
prematurely when a certain condition is met. Once break is 
encountered, the loop stops immediately, and control moves 
to the next statement after the loop

```python
for i in range(10):
    if i == 5:
        break
    print(i)
```

#### 🧠 Used in:

* Early stopping
* Search algorithms
* Error handling

***

### continue (Skip current iteration)

* The Continue statement skips one of the iteration and rest of 
the iterations will run for understanding lets say you will not 
run the 16th lap but all other lap will be there.
* Skips the rest of the current iteration and jumps back to the loop start

```python
for i in range(5):
    if i == 2:
        continue
    print(i)
```

#### 🧠 Used in:

* Skipping bad data
* Ignoring invalid rows
* Filtering

***

### pass (Do nothing, intentionally)

```python
for i in range(3):
    pass
    print(i)
```

#### Used when:

* Structure needed
* Logic to be added later
* Placeholder in pipelines
* Abstract pipelines

***

### Nested Loops (Loop inside loop)

```python
for i in range(2):
    for j in range(3):
        print(i, j)
```

***

### Loop Optimization

##### ❌ Bad practice

```python
for i in range(len(data)):
    print(data[i])
```

##### ✅ Better

```python
data=[3,55,7,1]
for item in data:
    print(item)
```

##### ❌ ML anti-pattern

```python
for row in data:
    for col in row:
        process(col)
```

**ML best practice**

* Use vectorization (NumPy)
* Batch processing
* Avoid Python loops when possible

***

### ML & Data Preprocessing Use Cases

#### 🔹 Epoch loop

```python
for epoch in range(epochs):
    train()
```

#### 🔹 Batch processing

```python
for batch in batches:
    update_weights(batch)
```

## DEEPER TOPICS RELATED TO LOOPS
### else with loops

```python
for i in range(5):
    if i == 10:
        break
else:
    print("Loop finished without break")
```

#### 🧠 Meaning:

* else runs only if loop completes normally
* Does NOT run if break happens

#### Used in:

* search algorithms
* validation logic

***

### enumerate() — index + value safely

```python
data = ["a", "b", "c"]

for index, value in enumerate(data):
    print(index, value)
```

***

### zip() — parallel looping

```python
names = ["A", "B"]
scores = [90, 80]

for n, s in zip(names, scores):
    print(n, s)
```

#### Used heavily in:

* feature-label pairing
* datasets

***

### List Comprehensions (LOOPS IN ONE LINE)

```python
squares = [x*x for x in range(5)]
print(squares)
```

```python
#Equivalent to:

squares = []
for x in range(5):
    squares.append(x*x)
    print(squares)
```

##### Used everywhere in:

* preprocessing
* feature creation

***

### Generator Expressions (MEMORY EFFICIENT)

```python
gen = (x*x for x in range(10))
print(gen)
```

Does not compute all at once.

##### Used in:

* large datasets
* streaming data
