#### What is Programming?
Programming = Giving instructions to a computer to solve problems automatically.

Programming is:
1. Breaking a problem into logical steps
2. Writing those steps in a language computers understand
3. Ensuring correctness, efficiency, and reusability

#### How Python Executes Code?
-> You write .py file

-> Python converts code → bytecode

-> Bytecode runs on Python Virtual Machine (PVM)

#### Interpreted vs Compiled: 
##### Compiled language (C++)

1. Code → Machine code → Run
2. Faster execution
3. Harder to debug

##### Interpreted language (Python)

1. Code executed line by line
2. Easier debugging
3. Slower speed

**#Important note :** Python is:

##### Both compiled (to bytecode) AND interpreted (by PVM)

### INPUT / OUTPUT & DEBUGGING MINDSET

#### INPUT()

What input() actually does

```python
name = input("Enter name: ")
```

##### Internally Python does:

* Pauses program execution
* Waits for keyboard text
* Reads characters as string
* Stores in variable

##### 👉 Important truth:
input() ALWAYS returns a string

**Even if user types:**

```python
25
```

Python stores: "25" as string

#### Multi-value input

```python
a, b = map(int, input().split())
```

##### What happens step-by-step

input → "10 20"

split → ["10","20"]

map int → [10,20]

unpack → a=10, b=20

This pattern is used in competitive coding & ML datasets.

#### OUTPUT
