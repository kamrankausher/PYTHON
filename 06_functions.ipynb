# Functions

### What is a function?

* A reusable named block of logic that transforms input → output.
* Functions in Python group reusable code into a block that 
can be executed by calling the function name. This helps 
avoid repetition and makes programs modular and readable
* There are many in-build functions in python like print(), input()
len() etc
* But we can create our own function and they are called as 
user defined functions. To make your own function we have 
to use def keyword and then name the function. After this 
we have to call the function using name() and paranthesis

##### Very important mental model:

data → function → new data

##### This is EXACTLY how ML works:

raw data → preprocessing → features → model → predictions

***

### PARAMETERS vs ARGUMENTS

#### Parameters (placeholders)

* Parameters are variables listed inside the function definition. 
* For making the function we have to accept inside the 
parenthesis of the function.

```python
def add(a, b):
    print(a + b)
```

a and b are parameters — empty variables waiting for values.

```python
def hello(name,age):             #name & age are Parameters
    print(f"your name is {name} and your age is {age}")
```

#### Arguments (actual values)

Arguments are the Values passed to a function when it is called

```python
hello("Kamran",22)       #Kamran & 22 are Arguments
```

```python
add(5, 3)
```

5 and 3 are arguments — real data.

#### Visual model

```python
function add(a,b)
            ↑ ↑
arguments → 5 3
```

Python creates new local variables:

```python
a = 5
b = 3
```

***

### Types of arguments

#### Positional arguments

Order matters

First parameter, will always capture first 
argument and so on. These arguments are called 
argument Positional arguments

**First Example**

```python
def add(a,b): 
    return a+b
print(add(3,5))        #3 is assigned to 'a' & 5 is assigned to 'b'
```

```python
def subtract(a,b):
    print(a-b)

subtract(10,5)  # 5              #10 is assigned to 'a' & 5 is assigned to 'b'
subtract(5,10)  # -5             #5 is assigned to 'a' & 10 is assigned to 'b'
```

* First example shows how positional arguments work

#### Keyword arguments

Order does NOT matter

```python
subtract(b=5, a=10)
```

**Second Example**

```python
def hello(name,age):             #name & age are Parameters
    print(f"your name is {name} and your age is {age}")
```

```python
hello(age = 22,name = "Kamran") 
```

* Second example shows how keyword argument works using this 
you can pass values in any order.

#### Default arguments

Provide fallback values.

**Third Example**

```python
def greet(name="User"):
    print("Hello", name)

greet()
greet("Kamran")
```

```python
def greet(name="guest"): 
    print(f"Hello ,{name}")
    
greet()                  #Use default value as guest
greet("Kamran")          # Use "Kamran"
```

### RETURN VALUES

#### Print vs Return

```python
def add(a,b):
    print(a+b)
```

This displays value but does NOT give it back.

Correct approach:

```python
def add(a,b):
    return a+b

result = add(2,3)
print(result)
```

#### Mental model

```python
print → talk to human
return → talk to program
```

ML pipelines use return, not print.

##### What return actually does?

* Sends value back
* Stops function execution immediately

```python
def test():
    return 5
    print("never runs")
```

### VARIABLE SCOPE

This causes many bugs.

#### Local scope

```python
def func():
    x = 10
    print(x)

func()
print(x)  # error
```

x only exists inside function

#### Global scope

```python
x = 10

def func():
    print(x)

func()
```

Function can read global variables.

#### Modifying global (dangerous)

```python
x = 10

def func():
    global x
    x = 20
```

Avoid in ML pipelines → leads to unpredictable behavior.
