#### What is an Operator? What is an Expression?

**Operator** → Symbol that performs an operation

**Expression** → Combination of values + operators that produces a result

x = 10 + 5


+ → operator

##### In Simple terms :
Operators are symbols that perform operations on variables 
and values. Python has several types of operators for 
different tasks like arithmetic, comparison, logical operations, 
and more

### Arithmetic Operators

* Used for numeric computation (core of ML)
* Arithmetic operators perform mathematical operations like 
addition, subtraction, multiplication, division, etc
* There are 7 types of arithmetic operator.

| Operator | Meaning        | Example       |
| -------- | -------------- | ------------- |
| `+`      | Addition       | `3 + 2 = 5`   |
| `-`      | Subtraction    | `5 - 2 = 3`   |
| `*`      | Multiplication | `4 * 2 = 8`   |
| `/`      | Division       | `5 / 2 = 2.5` |
| `//`     | Floor division | `5 // 2 = 2`  |
| `%`      | Modulus        | `5 % 2 = 1`   |
| `**`     | Power          | `2 ** 3 = 8`  |


##### 🧠 ML relevance

* Loss calculations
* Gradient updates
* Feature scaling
* Epoch & batch math

### Comparison Operators

* Comparison operators, also called relational operators, are 
used to compare two values
* Comparison operators will always provide Boolean result that 
is True and False
* Comparison operators will work with numbers but you can 
use them with strings as well.
* Strings will be comparing the Ascii values of string.
* comparison operators are as follows : 

| Operator | Meaning          |
| -------- | ---------------- |
| `==`     | Equal            |
| `!=`     | Not equal        |
| `>`      | Greater          |
| `<`      | Less             |
| `>=`     | Greater or equal |
| `<=`     | Less or equal    |


##### 🧠 Used in ML

* Early stopping
* Threshold checks
* Accuracy conditions

```python
#Equal
a = 3
b = 3
if a==b:
    print(True)
```

```python
#Not equal
A = 55
B = 6
if A!=B:
    print("Not Same")
```

```python
#Greater
n=55 
m=5
if n > m:
    print("n is Greater than m")
else:
    print("m is Greater than n")
```

```python
#Less
num_11= 20
mum_22= 44
if n < m:
    print("num_11 is Less than num_22")
else:
    print("num_22 is Less than num_11")
```

```python
#Greater or equal
E = 6000
F = 6000
if E >= F:
    print("Both are Equals")
else:
    print("E is greater than F")
```

### Logical Operators

* Logical operators in Python are used to combine multiple 
conditions and return a Boolean result (True or False)
*  Used to combine conditions.
* * and   - Return True if both condition are Tru2
* *  or      - Return True if at least one condition is True
* *  not    - Reverse the boolean value

| Operator | Meaning           |
| -------- | ----------------- |
| `and`    | Both true         |
| `or`     | At least one true |
| `not`    | Negation          |




```python
a = 10
b = 100
c = 20
if b> a and  b > c:
    print("b is bigger")
```

```python
if a > b or  c < b:
    print("B is Bigger no.")
```

### Operator Precedence

Order in which expressions are evaluated.

**🔝 High → Low**

*1.* ()

*2.* **

*3.* *, /, //, %

*4.* +, -

*5.* Comparison (<, >, ==)

*6.* not

*7.* and

*8.* or

```python
10 + 2 * 3   # 16, not 36
```

```python
(10 + 2) * 3       #Best practice:
```
