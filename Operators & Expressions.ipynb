#### What is an Operator? What is an Expression?

**Operator** ‚Üí Symbol that performs an operation

**Expression** ‚Üí Combination of values + operators that produces a result
Example:

x = 10 + 5


+ ‚Üí operator

#### Arithmetic Operators

Used for numeric computation (core of ML)

| Operator | Meaning        | Example       |
| -------- | -------------- | ------------- |
| `+`      | Addition       | `3 + 2 = 5`   |
| `-`      | Subtraction    | `5 - 2 = 3`   |
| `*`      | Multiplication | `4 * 2 = 8`   |
| `/`      | Division       | `5 / 2 = 2.5` |
| `//`     | Floor division | `5 // 2 = 2`  |
| `%`      | Modulus        | `5 % 2 = 1`   |
| `**`     | Power          | `2 ** 3 = 8`  |


##### üß† ML relevance

Loss calculations

Gradient updates

Feature scaling

Epoch & batch math

#### Comparison Operators

Used to compare values ‚Üí return True or False

| Operator | Meaning          |
| -------- | ---------------- |
| `==`     | Equal            |
| `!=`     | Not equal        |
| `>`      | Greater          |
| `<`      | Less             |
| `>=`     | Greater or equal |
| `<=`     | Less or equal    |


##### üß† Used in ML

Early stopping

Threshold checks

Accuracy conditions

```python
#Equal
a = 3
b = 3
if a==b:
    print(True)
```

```python
#Not equal
A = 55
B = 6
if A!=B:
    print("Not Same")
```

```python
#Greater
n=55 
m=5
if n > m:
    print("n is Greater than m")
else:
    print("m is Greater than n")
```

```python
#Less
num_11= 20
mum_22= 44
if n < m:
    print("num_11 is Less than num_22")
else:
    print("num_22 is Less than num_11")
```

```python
#Greater or equal
E = 6000
F = 6000
if E >= F:
    print("Both are Equals")
else:
    print("E is greater than F")
```

#### Operator Precedence

Order in which expressions are evaluated.

*üîù High ‚Üí Low*

*1.* ()

*2.* **

*3.* *, /, //, %

*4.* +, -

*5.* Comparison (<, >, ==)

*6.* not

*7.* and

*8.* or

```python
10 + 2 * 3   # 16, not 36
```

```python
(10 + 2) * 3       #Best practice:
```
