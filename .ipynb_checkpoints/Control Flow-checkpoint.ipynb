# Control Flow (if, elif, else)

#### What is Control Flow? (First principles)

Control flow determines:

* Which code runs
* Under what condition
* In what order

Without control flow:

* Programs would always run line-by-line
* No decisions
* No intelligence

**ðŸ‘‰ AI systems are nothing but complex decision systems**

*** 

### if Statement
ðŸ”¹ Syntax

```python
if condition:
    code_block
```

```python
loss = 0.05

if loss < 0.1:
    print("Model is learning")
```

##### ðŸ§  Important rule

* Condition must evaluate to boolean
* Indentation defines scope (very important in Python)

**Examples**

```python
n = "Sum"
m = "Sum"
if n==m:
    print("Both are same")
```

```python
loss= 0.2
accuracy = 0.9
if loss < 0.4 and accuracy > 0.7:
    print("Model is performing well")
```

***

### if-else

Used when exactly one path must run.

```python
if accuracy > 0.9:
    print("Good model")
else:
    print("Needs improvement")
```

**ðŸ‘‰ Either if OR else will execute, never both.**

***

### if-elif-else (Multiple decisions)

Used when you have multiple mutually exclusive conditions

```python
score = 87
if score >= 90:
    grade = "A"
elif score >= 75:
    grade = "B"
elif score >= 60:
    grade = "C"
else:
    grade = "Fail"
print(grade)
```

##### ðŸ§  Execution rule

* Checked top to bottom
* First True block executes
* Rest are skipped

***

### Nested Conditions (Conditions inside conditions)

```python
model_trained =True
if model_trained:
    if accuracy > 0.9:
        print("Deploy model")
    else:
        print("Tune hyperparameters")
```

***

### Truthy & Falsy Values

**ðŸ”» Falsy values in Python**

These evaluate to False:
* False
* None
* 0
* 0.0
* "" (empty string)
* [] (empty list)
* {} (empty dict)
* set()
  
 ***Everything else is Truthy.***

**ðŸ§  Example**

```python
data = []

if data:
    print("Data exists")
else:
    print("No data")
```

Used heavily in:

* Data validation
* ML pipelines
* API checks

***

### Real Decision Logic in ML Systems

**ðŸ”¹ Example 1: Early Stopping**

```python
if loss < threshold and epoch > min_epoch:
    stop_training = True
```

**ðŸ”¹ Example 2: Data Validation**

```python
if not data:
    raise ValueError("Dataset is empty")
```

**ðŸ”¹ Example 3: Model Selection**

```python
if accuracy > best_accuracy:
    best_model = current_model
```

**ðŸ”¹ Example 4: Inference Logic**

```python
if confidence > 0.8:
    return "Accepted"
else:
    return "Rejected"
```
