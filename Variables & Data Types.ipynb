# Variables, Data Types & Python Memory Model

## VARIABLES

### What is a Variable?

1. A variable is a name (label) that points to an object in memory
2. Python does not store values inside variables.
   
->  It stores objects in memory, and variables reference them

For example :
x = 10

10 ‚Üí object in memory (heap)

x ‚Üí reference (name) pointing to that object

## Mutability vs Immutability

### üîí Immutable types

int

float

bool

string

tuple

You cannot change the object itself.

```python
x = 10
x = x + 1   # new object created
```

### üîì Mutable types (preview)

list

dict

set

```python
lst = [1, 2, 3]
lst.append(4)   # same object modified
```

‚ö†Ô∏è This is where ML bugs are born.

### Why This Matters

#### üöÄ ML Performance

Object creation is expensive

Immutable objects create new memory

Vectorized operations avoid overhead

##### üêõ Bug-Free Code

Example bug:

```python
a = [1, 2]
b = a
b.append(3)
```

Now:

```python
a = [1, 2, 3]  # unexpected
```

This destroys ML pipelines if misunderstood.
